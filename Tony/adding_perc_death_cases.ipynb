```python
import pandas as pd
import numpy as np
```

```python
file = '../data/covid_with_lat_long.csv'
covid_df = pd.read_csv(file)
```

```python
covid_df.head()
```

```python
covid_df['cases as percent of population'] = covid_df['cases']/covid_df['population']
covid_df['deaths as percent of population'] = covid_df['deaths']/covid_df['population']
covid_df.head()
```

```python
covid_df = covid_df.drop(columns=['Unnamed: 0'])
covid_df.head()
```

```python
covid_df.to_csv(r'../data/final_data.csv')
```
